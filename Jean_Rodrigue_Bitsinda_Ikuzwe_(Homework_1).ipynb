**(i)** The solution of the schrodinger wave equation for a free particle constrained (-1,1)

```python
#Setting the grid and step size
N = 100
a = -1
b = 1
x = np.linspace(a,b,N)
h = x[1] - x[0]

#At the boundaries, the potential function V(x)=0, so the kinetic energy matrix T becomes

T = np.zeros((N-2,N-2))
for i in range(N-2):
    for j in range(N-2):
        if i==j:
            T[i,j] = 1/h**2
        elif np.abs(i-j)==1:
            T[i,j] = -1/(2*h**2)
        else:
            T[i,j] = 0
            
#Now From definition of hamiltonian operator H is defined as H = T + V, where V(x)=0
H = T

#Looking for the ground state and first excited state eigenvalues and eigenstates

val,vec = np.linalg.eig(H)
n = np.argsort(val)
n = n[0:2]
E=(val[n])
print("The energy states n are given as ", n)
print("The corresponding energies are ", E)
```

```python
'''Looking for the normalised ground state and first excited state eigenvalues 
  and eigenstates'''

val,vec = np.linalg.eig(H)
n = np.argsort(val)
n = n[0:2]
E=(val[n]/(val[n][0]))
print("The normalised energy states n are: ", n)
print("The corresponding energies are given as ", E)
```

```python
#plotting wavefunctions for the ground state and first excited state

plt.figure(figsize=(12,7))
for i in range(len(n)):
    y1 = []
    y1 = np.append(y1,vec[:,n[i]])
    y1 = np.append(y1,0)
    y1 = np.insert(y1,0,0)
    plt.plot(x,y1,lw = 3, label="n = {} ".format(i))
    plt.xlabel('x', size = 20)
    plt.ylabel('$\psi$(x)',size = 20)
plt.legend()
plt.title('The plot of the ground state and first excited state wavefunctions',size = 20)
plt.show()
```
